## 1. 보험 상품의 보험료 및 보상 분석

 ##### <주어진 데이터프레임을 'PROD_ABBR' 열을 기준으로 그룹화한 후, 각 그룹에서 'NB_WRTN_PREM_AMT'와 'PRD_INCRD_LOSSES_AMT' 열의 값을 합산하는 것을 수행>
 ##### 이후에는 결과를 새로운 데이터프레임으로 변환하여 출력

#### 코드내용

#### 각 보험 상품('PROD_ABBR'= product Abbreviation) 별로 작성된 보험료('NB_WRTN_PREM_AMT')와 발생한 손해('PRD_INCRD_LOSSES_AMT')의 총액을 알 수 있음.
#### 'NB_WRTN_PREM_AMT'는 해당 보험 상품에 대한 보험료의 총액을 나타냄./ 'PRD_INCRD_LOSSES_AMT'는 해당 보험 상품에 대한 발생한 손해의 총액을 나타냄
#### <이를 통해 각 보험 상품의 매출과 손해를 비교하여 수익성을 평가>
#### 출력된 데이터프레임인 result1은 각 보험 상품별로 총 보험료와 총 보상을 보여주는 표
####  'PROD_ABBR' 열은 보험 상품의 코드를 나타내고, 'NB_WRTN_PREM_AMT'와 'PRD_INCRD_LOSSES_AMT' 열은 각각
#### 해당 보험 상품의 총 보험료와 총 보상을 나타냄.

```python

result1 = df.groupby('PROD_ABBR').agg({'NB_WRTN_PREM_AMT': 'sum', 'PRD_INCRD_LOSSES_AMT': 'sum'}).reset_index()
print("1. 보험 상품의 보험료 및 보상 분석")
print(result1)
```

```python
# 보험료가 가장 많은 상품 찾기
max_premium_product = df.loc[df['NB_WRTN_PREM_AMT'].idxmax()]['PROD_ABBR']

# 보상이 가장 많이 발생한 상품 찾기
max_loss_product = df.loc[df['PRD_INCRD_LOSSES_AMT'].idxmax()]['PROD_ABBR']

print("가장 많은 보험료를 낸 상품:", max_premium_product)
print("가장 많은 보상이 발생한 상품:", max_loss_product)
```

#### PROD_ABBR

1. ANNIV: Anniversary Insurance (기념일 보험) - 보험 가입일 기준으로 정해진 주기마다 보상을 제공하는 보험 상품입니다.
2. BOILERMACH: Boiler Machinery Insurance (보일러 및 기계 보험) - 공장이나 건물에 설치된 보일러 및 기계의 손해에 대한 보험입니다.
3. BOP: Business Owner's Policy (사업자 보험 정책) - 소기업 소유자를 위한 포괄적인 보험 패키지로, 일반적으로 재산 및 책임 보험을 포함합니다.
4. COMMAUTO: Commercial Automobile Insurance (상업용 자동차 보험) - 비즈니스 목적으로 사용되는 자동차에 대한 보험입니다.
5. COMMINLMAR: Commercial Inland Marine Insurance (상업용 내륙해양 보험) - 상업적으로 사용되는 물품의 운송 중 발생할 수 있는 손해에 대한 보험입니다.
6. COMMPOL: Commercial Pollution Liability Insurance (상업용 오염 책임 보험) - 비즈니스가 환경 오염으로 인한 손해에 대해 보호받는 보험입니다.
7. COMMUMBREL: Commercial Umbrella Insurance (상업용 우산 보험) - 다른 책임 보험의 한도를 초과하는 금액에 대한 추가적인 보호를 제공하는 상업 보험입니다.
8. CRIME: Crime Insurance (범죄 보험) - 사기, 절도, 위조 등 범죄로 인한 손해에 대한 보상을 제공하는 보험입니다.
9. CYCLES: Cycles Insurance (오토바이 보험) - 오토바이에 대한 보험입니다.
10. DTALK: 이에 대한 추가 정보가 없으므로 정확한 의미는 파악하기 어렵습니다.
11. DWELLFIRE: Dwelling Fire Insurance (주택 화재 보험) - 주택에 발생하는 화재에 대한 보상을 제공하는 보험입니다.
12. FIREALLIED: Fire and Allied Lines Insurance (화재 및 기타 연계 보험) - 화재, 번개, 폭발, 운송 중 손실 등에 대한 보상을 제공하는 보험입니다.
13. GARAGE: Garage Insurance (차고 보험) - 차고 운영에 관련된 위험에 대한 보험입니다.
14. GENERALIAB: General Liability Insurance (일반 책임 보험) - 사업 또는 개인이 제3자로부터 손해배상 청구를 받을 경우를 대비한 보험입니다.
15. HOMEOWNERS: Homeowners Insurance (주택 소유자 보험) - 주택 소유자를 위한 재산 및 책임 보호를 제공하는 보험입니다.
16. MOBILEHOME: Mobile Home Insurance (이동식 주택 보험) - 이동식 주택에 대한 보험입니다.
17. MOTORHOM12: Motor Home Insurance (모터홈 보험) - 모터홈에 대한 보험입니다.
18. PERSAIP: Personal Automobile Insurance Policy (개인용 자동차 보험 정책) - 개인이 소유한 자동차에 대한 보험입니다.
19. PERSINLMAR: Personal Inland Marine Insurance (개인 내륙해양 보험) - 개인용 보트나 보트 운송 중 발생할 수 있는 손해에 대한 보험입니다.
20. SNOWMOBI12: Snowmobile Insurance (스노모바일 보험) - 스노모바일에 대한 보험입니다.
21. WORKCOMP: Workers' Compensation Insurance (근로자 보상 보험) - 직원이 업무 수행 중 발생한 사고에 대한 보상을 제공하는 보험입니다.
22. YACHT: Yacht Insurance (요트 보험) - 요트에 대한 보험입니다.
